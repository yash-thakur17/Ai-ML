```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

number = int(input("Enter a number so you can see number of data: "))
Store_Data = pd.read_csv("SampleSuperstore.csv", nrows=number)

# Find a city
City_Name= input("Enter a which city you can Find: ")
Same_city = Store_Data[Store_Data['City'] == City_Name ]
print(len(Same_city))
print(Same_city)

# Find State name
State_name = input("Enter a which State you can see: ")
state_name = Store_Data[Store_Data['State'] == State_name]
print(len(state_name))
print(state_name)

plt.figure(figsize=(12,6))
plt.subplots_adjust(hspace=0.4, wspace=0.4)


# graph for 3 categorys ho much sales
def Graph1():
    plt.subplot(2, 2, 1)
    Total_Category = Store_Data.groupby('Category')['Sales'].sum()
    plt.plot(Total_Category.index, Total_Category.values, label="Sales", color='blue', marker='o')
    plt.title("Sales Details by Category")
    plt.xlabel("Category Name")
    plt.ylabel("Total Sales")
    plt.legend()
    plt.grid()

def Graph2():
    plt.subplot(2, 2, 2)
    Total_Category = Store_Data.groupby('Category')['Sales'].sum()
    plt.bar(Total_Category.index, Total_Category.values, label="Sales", color='blue')
    plt.title("Sales Details by Category")
    plt.xlabel("Category Name")
    plt.ylabel("Total Sales")
    plt.legend()

def Graph3():
    plt.subplot(2,2,3)
    values = Total_Category.values
    labels = Total_Category.index 
    plt.pie(values,labels=labels,colors=['blue','green','red'], autopct='%1.1f%%',startangle=90)
    plt.title('Pass vs Fail Students')
    plt.legend()
    plt.grid()

def Graph4():
    plt.subplot(2,2,4)
    sns.boxenplot(x=Store_Data['Sales'])
    plt.title("The total sales Distributions: ")

chooice = int(input("Enter a chooice to show a graph . 1.Graph\n2.Bar Chart\n 3.pie chart\n4.Distribution"))

if chooice == 1:
    print("Graph 1 ")
    Graph1()
elif chooice == 2:
    print("Graph 2")
    Graph2()
elif chooice == 3:
    print("Graph 3")
    Graph3()
elif chooice == 4:
    print("Graph 4")
    Graph4()
elif chooice == 5:
    print("The Chooic are exited")
    exit
else:
    print("Pleses Enter a Valid chooice in a  1.Graph\n2.Bar Chart\n 3.pie chart\n4.Distribution")
# Enter a funiture name wish to view:
Category_c = input("Enter type of furniture you wish to view and wich city or State: ")
Category_name = Store_Data[Store_Data['Category'] == Category_c]
print(Category_name)

filtered_Data = Store_Data[
    (Store_Data['Category'] == Category_c)
]

# Enter a Sub-category to see it
Sales_s = input("Enter a category to you find a total: ")
Sales_name = Store_Data[Store_Data['Sub-Category'] == Sales_s]
print(Sales_name)

filtered_Data = Store_Data[
    (Store_Data['Sub-Category'] == Sales_s)
]

# Total of all sub-category
total_of_Subcategory = filtered_Data['Sales'].sum()
print(("Choose of a your sub-category sales is: "),total_of_Subcategory)

# Total of all Sales
Total_Sales = Store_Data['Sales'].sum()
print(("The Total sum is: "),Total_Sales)

# Find a mean of Sales
Mean_sales = Store_Data['Sales'].mean()
print(("Mean of Sales Is: "),Mean_sales)

# Normallize a Profite
Store_Data['Normallize_revice'] = Store_Data["Profit"] / Store_Data["Profit"].max()
print(("\nThe Max of normallization marks: "),Store_Data['Normallize_revice'].max())


Store_Data.head()
plt.tight_layout()
plt.show()
```
